# - Plotly ile Kodlama

Plotly ktphanesinin syntax dier grsel ktphanelere gre karmak gelebilir. Ancak aslnda mant anlaldnda tpk bir boluk dolurur gibi kod paras oluturabiliriz. Bunun iin ncelikle plotly ktphanesinin Figure objesini kavramak gerekmektedir. Figure objesi genellikle iki ana bileenden oluur. Bunlar Data ve Layouttur.

* Data bileeni, grafiin izilmesi iin gereken veriyi ierirken
* Layout bileeni grafiin grnm ile ilgili ayarlar iermektedir.

#### Data: 

Data bileeni, genellikle bir veya birden fazla izgi (trace) ierir. Her bir izgi, Scatter, Bar, Box, Heatmap vb. olabilir ve her biri kendi zel ayarlarna sahiptir.

#### Layout:

Layout bileeni, grafiin bal, eksen balklar, zgara izgileri, arka plan rengi vb. gibi eitli stil ve formatlama ayarlarn ierir.



```python
import plotly.graph_objects as go

# Data
trace = go.Scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13])

# Layout
layout = go.Layout(title='Basit izgi Grafii', xaxis=dict(title='X Ekseni'), yaxis=dict(title='Y Ekseni'))

# Figure
fig = go.Figure(data=[trace], layout=layout)

# izim
fig.show()
```

## Line Chart

Line grafik (izgi grafii), genellikle zamanla deien deerleri gstermek iin kullanlan bir grafik trdr. Line grafik zamanla deiimi gstermek, trendleri belirlemek, karlatrmalar yapmak, ngrlerde bulunmak iin de kullanlr.

Plotly ktphanesi ile bir line chart izdirelim. Bu kod paras ile, Citation and Teaching vs World Rank of Top 100 Universities bal altnda, dnya sralamas (x ekseninde) ile atf says ve retim skoru (y ekseninde) arasndaki ilikiyi gsteren iki izgi grafii izdirilecektir. Her izgi, farkl bir renkte olacak, ve fareyle her bir nokta zerine gelindiinde, o noktaya karlk gelen niversitenin ad grntlenecektir.

```python
#plt.style.use("ggplot")
# creating trace 1

trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = "lines",    # Bu dalm izi iin izim modunu belirler. "mode=text" ise text'i yazdrr "text" eleri koordinatlarda grnr. 
                                       # Aksi takdirde, fareyle zerine gelindiinde "text" eleri grnr. 
                                       # 20'den az nokta varsa ve iz ynlanmamsa varsaylan deer "lines+markers"dir. Aksi takdirde, "lines".
                    name = "citations",
                    marker = dict(color = "rgb(16, 112, 2)"), # {"color" : "rgb(16, 112, 2)"}
                    text = df.university_name)
# creating trace 2
trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines + markers",
                    name = "teaching",
                    marker = dict(color = "rgb(80,26,80)",
                                  size = 10,
                                  symbol = "star",
                                  line = dict(
                                         color = "yellow",
                                         width = 1),),
                    text = df.university_name)

data = [trace1, trace2]

layout = dict(title = "Citation and Teaching vs Work Rank of Top 100 Universities", #axes olarak dnebiliriz
              xaxis = dict(title = "World Rank", ticklen = 5, tickcolor = "crimson", zeroline = False),
              yaxis = dict(title = "Value", ticklen = 5, tickcolor = "crimson", zeroline = False))

fig = dict(data = data, layout = layout)  #figure iine yukarda tanmladmz grafik deikenlerini ve layout parametrelerini dict olarak verip 
iplot(fig)                                #iplot(fig) ile izdiriyoruz.
```

### Scatter Plot

Scatter plot (salm grafii), iki deiken arasndaki ilikiyi, veri dalmn ve veri gruplarn grsel olarak ayrt etmek iin kullanlan bir grafik trdr.

imdi plotly ile bir scatter plot izdirelim. Bu kod paras ile 2014, 2015 ve 2016 yllarnda ilk 100 niversitenin citation (atf) saylarna kar dnya sralamalarn bir scatter plot ile grselletirilmesi hedeflenmitir.

```python
import plotly.graph_objs as go
from plotly.offline import iplot

# Veri erevesini kullanarak trace'leri oluturun
trace1 = go.Scatter(
                    x = timesData[timesData['year'] == 2014]['world_rank'],
                    y = timesData[timesData['year'] == 2014]['citations'],
                    mode = "markers",
                    name = "2014",
                    marker = dict(color = "rgb(255,128,255)"),
                    text = timesData[timesData['year'] == 2014]['university_name'])

trace2 = go.Scatter(
                    x = timesData[timesData['year'] == 2015]['world_rank'],
                    y = timesData[timesData['year'] == 2015]['citations'],
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = "rgb(255,128,2)"),
                    text = timesData[timesData['year'] == 2015]['university_name'])

trace3 = go.Scatter(
                    x = timesData[timesData['year'] == 2016]['world_rank'],
                    y = timesData[timesData['year'] == 2016]['citations'],
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = "green"),
                    text = timesData[timesData['year'] == 2016]['university_name'])

data = [trace1, trace2, trace3]

layout = dict(title = "Citation vs Work Rank of Top 100 Universities with 2014, 2015, 2016",
              title_x = 0.5, title_y = 0.9,
              xaxis = dict(title = "World Rank"),
              yaxis = dict(title = "Citation"))

fig = dict(data = data, layout = layout)

# Grafii grselletirin
iplot(fig)
```

### - Pie Chart
Pie chart (dilim grafii), bir btnn belirli paralarnn oransal dalmn gstermek, kategorik verinin grselletirilmesi, bir btn iindeki ksmlarn gsterilmesi iin kullanlmaktadr. Ayrca Pie chartlar, bir toplamn (rnein bir btenin, nfusun veya satlarn) belirli blmlerinin ne kadarn oluturduu hakknda bilgi almak iin uygundur.

Bunun haricinde Pie chart kullanrken dikkat edilmesi gereken birka nemli nokta vardr. Bunlar;

* Az Sayda Kategori: Pie chartlar genellikle az sayda kategori iin en iyisidir. ok fazla kategori varsa, grafik okunmas zorlaabilir.

* Yzdeliklerin Toplam: Dilimlerin yzdelikleri toplam %100 olmaldr.

* Etiketleme: Pie chartlarda dilimlerin neyi temsil ettiini belirten etiketlerin veya aklamalarn net ve okunabilir olmas nemlidir.

imdi plotly ile bir pie chart izdirelim. Bu kod paras ile belirli niversitelerin renci saylar bir pasta grafii ile gsterilmektedir. Her bir dilim bir niversiteyi temsil eder. Dilimlerin zerine gelindiinde ilgili niversitenin ismi ve o dilimin yzdesi grntlenir. Ayrca, son iki niversite dilimi merkezden ayrlarak vurgulanmtr ve grafik donut eklinde bir bolua sahiptir.

```python
# nce 2016 ylna ait verileri ve ilk 7 kayd alyoruz.
df2016 = timesData[timesData.year == 2016].iloc[:7,:]

# num_students stununu alp, virglleri kaldryoruz ve stringi float'a eviriyoruz.
df = df2016.num_students
df1 = (df.str.replace(",", "")).astype("float")
df2 = list((df.str.replace(",", "")).astype("float"))

# List comprehension ile num_students stunundaki deerleri floata eviriyoruz.
pie_list = [float(each.replace(",", "")) for each in df2016.num_students]

# labels deikeni ile pasta grafii dilimlerine isimlerini yerletiriyoruz.
labels = df2016.university_name

# Figure objesini oluturuyoruz. 
fig = {
    "data" : [
        {
            "values" : pie_list,  # Yzdelik dilimleri belirleyecek olan deerler.
            "labels" : labels,  # Her bir dilimin etiketi.
            "domain" : {"x": [0, .5]},  # Grafikteki konumu.
            "name" : "Number Of Students Rates",  # Grafik ismi.
            "hoverinfo" : "label + percent",  # Fareyle zerine gelindiinde gsterilecek bilgi.
            "hole" : .2,  # Ortadaki boluun bykl, bu bir donut grafii yapar.
            "type" : "pie",  # Grafik tr.
            "pull" : [0,0,0,0,0, 0.2, 0.2]  # Belirli dilimleri vurgulamak iin ekiyoruz, burada son iki dilim ekilmi.
        },
    ],
    "layout" : {
        "title" : "Universites Number of Students rates",  # Ana balk.
        "annotations" : [
            {"font" : {"size" : 20},  # Font bykl.
             "text" : "Number of Students",  # Yorum yazs.
             "x" : 0.3,  # X konumu.
             "y" : 1  # Y konumu.
            },
        ]
    }
}

# Oluturulan Figure objesini izdiriyoruz.
iplot(fig)
```

### - Buble Chart
Bubble chart (baloncuk grafii), genellikle  boyutlu veriyi bir arada gstermek iin kullanlan bir grafik trdr.

Bubble chart kullanmnn ana nedenleri:

*  Boyutlu Veriyi Gstermek: X ve Y eksenlerinde iki deikeni grselletiren bir scatter plotun geniletilmi bir versiyonudur. Ancak, her veri noktas bir baloncuk (bubble) ile temsil edilir ve bu baloncuun bykl nc bir deikeni gsterir.

* Deerler Arasndaki likiyi Gstermek: X ve Y eksenleri arasndaki ilikiyi, ayn zamanda nc bir deikenin bykln de gz nnde bulundurarak gsterir.

* Ekstra Bilgi Salamak: Baloncuklarn renkleri, bir drdnc deikeni gstermek iin de kullanlabilir, bylece daha fazla bilgi bir arada sunulabilir.

Bubble chart kullanrken dikkat edilmesi gereken baz noktalar:

* Okunabilirlik: ok fazla baloncuk olduunda veya baloncuklar birbirine ok yakn olduunda, grafik karmak ve okunmas zor hale gelebilir.

* Boyut Yanltc Olabilir: ki baloncuun alanlarn karlatrrken, byklkler arasndaki farklar insan gz tarafndan yanltc olabilir.

* Etkili Etiketleme: zellikle birok baloncuk olduunda, hangi baloncuun hangi veriye ait olduunu belirten etiketlerin net ve okunabilir olmas gerekir.

zetlemek gerekirse,  veya daha fazla deikenin ilikisini bir arada grselletirmek iin bubble chart kullanlr.

imdi plotly ile bir bubble chart izdirelim. Bu kod paras ile niversitelerin dnya sralamas X ekseni zerinde, teaching skorlar ise Y ekseni zerinde gsterilmektedir. Her bir nokta, bir niversiteyi temsil etmekte, noktalarn renkleri international skorlarna, boyutlar ise renci saysna gre deimektedir. Bu grafik, eitli niversitelerin teaching ve international skorlarnn karlatrlmasn gstermektedir.

```python
df2016 = timesData[timesData.year == 2016].iloc[:20,:]  # 2016 ylnda dnyann en iyi 20 niversitesini se

num_students_size  = [float(each.replace(",", ".")) for each in df2016.num_students]  # num_students kolonunu float tipine evir

international_color = [float(each) for each in df2016.international]  # international kolonunu float tipine evir

data = [  # Scatter plot oluturulacak data'y tanmla
    {
        "y" : df2016.teaching,  # y ekseni olarak teaching kolonunu kullan
        "x" : df2016.world_rank,  # x ekseni olarak world_rank kolonunu kullan
        "mode" : "markers",  # scatter plot iin markers modunu kullan
        "marker" : {  # marker zelliklerini ayarla
            "color" : international_color,  # marker rengini international_color listesiyle belirle
            "size" : num_students_size,  # marker boyutunu num_students_size listesiyle belirle
            "showscale" : True  # renk leini gster
        },
        "text" : df2016.university_name  # her marker iin niversite ismini gster
    }
]
iplot(data)  # grafii iz
```

### - Histogram
Histogram, bir veri setindeki deerlerin dalmn gstermek iin kullanlan bir grafik trdr.

Histogram kullanmnn ana nedenleri:

* Veri Dalmn Gstermek: Histogram, veri setindeki deerlerin frekans dalmn gsterir. Yani, belirli aralklardaki deerlerin ne kadar sk olduunu gsterir.

* Merkezi Eilimi ve Dalm Deerlendirmek: Histogram araclyla bir veri setinin ortalamas, modu, medyan gibi merkezi eilim ltlerini ve varyans, standart sapma gibi dalm ltlerini grsel olarak deerlendirebilirsiniz.

* Normal Dalm Gzlemlemek: Veri setinin normal bir dalma sahip olup olmadn anlamak iin histogram kullanlabilir.

* Aykr Deerleri Tespit Etmek: Histogram zerinde beklenenden farkl ekillerde veya beklenmeyen yerlerde oluan kntlar, aykr deerlerin ya da zel durumlarn varlna iaret edebilir.

Histogram kullanrken dikkat edilmesi gereken baz noktalar:

* Bin Seimi: Histogramda bin ad verilen stunlarn genilii nemlidir. ok fazla bin kullanlrsa, veri setindeki doal varyasyon belirginleir; ok az bin kullanlrsa, veri setinin yaps kaybolabilir.

* Dzgn Dalm: Tm binglerin ayn genilikte olmas, veri dalmn doru bir ekilde yorumlamay kolaylatrr.

zetle, bir veri setinin frekans dalmn grselletirmek, merkezi eilim ve dalm deerlendirmek iin histogram kullanlr.

imdi plotly ile bir histogram izdirelim. Bu kod paras ile 2011 ve 2012 yllar arasnda niversitelerdeki renci-personel orannn dalmn gsteren iki histogram izdirlmitir. Histogramlar st ste geldii iin, opacity parametresi ile effaflklar ayarlanmtr. Bu sayede, iki yl arasndaki farkllklar ve benzerlikler daha net bir ekilde gzlemlenebilir.

```python
# 1. 2011 ylna ait renci-personel orann x2011 deikenine atar.
x2011 = timesData.student_staff_ratio[timesData.year == 2011]

# 2. 2012 ylna ait renci-personel orann x2012 deikenine atar.
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

# 3. 2011 yl iin bir histogram oluturur.
trace1 = go.Histogram(
    x=x2011,  # X eksenine x2011 deikenini atar.
    opacity=0.75,  # Histogramn effafln belirler.
    name = "2011",  # Histogramn adn belirler.
    marker=dict(color='rgba(171, 50, 96, 0.6)')  # Histogramn rengini belirler.
)

# 4. 2012 yl iin bir baka histogram oluturur.
trace2 = go.Histogram(
    x=x2012,  # X eksenine x2012 deikenini atar.
    opacity=0.75,  # Histogramn effafln belirler.
    name = "2012",  # Histogramn adn belirler.
    marker=dict(color='rgba(12, 50, 196, 0.6)')  # Histogramn rengini belirler.
)

# 5. ki histogram bir veri listesi iine alr.
data = [trace1, trace2]

# 6. Grafiin dzenini ve balklarn ayarlar.
layout = go.Layout(
    barmode='overlay',  # ki histogramn st ste gelmesini salar.
    title=' students-staff ratio in 2011 and 2012',  # Grafiin baln belirler.
    xaxis=dict(title='students-staff ratio'),  # X ekseninin baln belirler.
    yaxis=dict(title='Count')  # Y ekseninin baln belirler.
)

# 7. Veri ve dzeni kullanarak bir Figure nesnesi oluturur.
fig = go.Figure(data=data, layout=layout)

# 8. Oluturulan Figure nesnesini izer.
iplot(fig)
```

### - Box Plots
Box plot, bir veri setinin zet istatistiklerini (medyan, eyreklikler, aykr deerler vb.) gstermek iin kullanlan bir grafik trdr.

Box plot kullanmnn ana nedenleri:

* Veri Dalmn Gstermek: Box plot, veri setinin merkezi eilimini, dalmn ve aykr deerlerini (outliers) gstererek veri dalmnn bir zetini sunar.

* Karlatrmalar Yapmak: Farkl kategoriler veya gruplar arasndaki veri dalmlarn yan yana koyarak karlatrmay kolaylatrr.

* Aykr Deerleri Tespit Etmek: Box plot, veri setindeki potansiyel aykr deerleri belirgin bir ekilde gsterir.

* Dalmn eklini Deerlendirmek: Verinin simetrik olup olmad, dalmn ne kadar skk veya yayld gibi bilgileri grsel olarak salar.

Box plot kullanrken dikkate alnmas gereken hususlar:

* Veri Yaps: Box plot, srekli (kantitatif) veri iin uygundur. Kategorik (nitel) veri ile birlikte kullanldnda, her kategori iin ayr bir box plot oluturulabilir.

* Veri Dalmnn Anlalmas: Box plot, medyan, birinci ve nc eyreklikler gibi zet istatistikleri gsterir, ancak veri dalmnn tam detaylarn (rnein, mod veya bimodal bir dalm olup olmad) gstermez.

zetle, bir veri setinin zet istatistiklerini grselletirmek, aykr deerleri tespit etmek ve farkl gruplar veya kategoriler arasnda karlatrmalar yapmak iin box chart kullanlr.

imdi plotly ile bir box plot izdirelim. Bu kod paras ile 2015 ylnda niversitelerin toplam puanlar (total_score) ve aratrma puanlar (research) iin iki adet Box Plot grafii izeceiz. Bu grafikle elde ettiimiz Box Plot'lar ile verilerin zet istatistikleri (medyan, eyrekler, minimum, maksimum ve aykr deerler) hakknda grsel bir ekilde bilgi edineceiz.

```python
# 1. 2015 ylna ait verileri x2015 deikenine atar.
x2015 = timesData[timesData.year == 2015]

# 2. x2015.total_score verilerini kullanarak bir Box Plot oluturur.
trace0 = go.Box(
    y=x2015.total_score,  # Box Plot'un y ekseninde gsterilecek veriyi belirtir.
    name = 'total score of universities in 2015',  # Grafiin adn belirtir.
    marker = dict(color = 'rgb(12, 12, 140)')  # Grafiin rengini belirtir.
)

# 3. x2015.research verilerini kullanarak bir Box Plot oluturur.
trace1 = go.Box(
    y=x2015.research,  # Box Plot'un y ekseninde gsterilecek veriyi belirtir.
    name = 'research of universities in 2015',  # Grafiin adn belirtir.
    marker = dict(color = 'rgb(12, 128, 128)')  # Grafiin rengini belirtir.
)

# 4. Oluturulan Box Plot'lar bir liste ierisinde toplar.
data = [trace0, trace1]

# 5. Oluturulan Box Plot'lar izer.
iplot(data)
```

### - Scatter Matrix Plots
Scatter Matrix Plots (pair plots olarak da adlandrlr), birden ok srekli deikenin birbiriyle olan ilikisini incelemek iin kullanlan bir grselletirme yntemidir.

Scatter Matrix Plots kullanmnn ana nedenleri:

* Deikenler Aras likileri Gzlemlemek: Birden fazla deiken arasndaki potansiyel ilikileri veya korelasyonlar hzl bir ekilde grselletirmek iin kullanlr.

* Veri Dalmn Gstermek: Her deikenin kendisi iin dalmn (genellikle histogram veya kernel density estimation eklinde) ana kegen zerinde gsterir.

* Multivaryant Veri Analizi: Scatter matrix, multivaryant veri setlerindeki deikenlerin birbiriyle olan etkileimlerini anlamaya yardmc olabilir.

* Aykr Deerleri ve Veri Younluklarn Tespit Etmek: Deikenlerin birbiriyle olan ilikisini gsteren scatter plotlarda, aykr deerler veya zel veri younluklar gzlemlenebilir.

Scatter Matrix Plots kullanrken dikkate alnmas gereken hususlar:

* Veri Younluu: ok fazla gzlem veya deiken olduunda, scatter matrix plots youn ve okunmas zor olabilir. Bu durumda, rnekleme veya daha spesifik deiken seimi yapmak faydal olabilir.

* Kategorik Deikenler: Bu tr plotlar genellikle srekli deikenler iin daha uygundur. Kategorik deikenlerin dahil edilmesi durumunda, farkl renklendirme veya simgeler kullanarak grselletirme yaplabilir.

zetle, birden fazla srekli deiken arasndaki ilikileri hzla grselletirmek, deikenlerin dalmlarn incelemek ve multivaryant veri analizi yapmak iin Scatter Matrix Plots kullanlabilir.

imdi plotly ile bir Scatter Matrix Plots izdirelim. Bu kod paras ile 2015 ylndaki niversitelerin research, international, ve total_score stunlarna dayal olarak bir scatter plot matrisi oluturur. Scatter plot matrisi, birden ok deiken arasndaki ilikileri gsteren bir seri scatter plot ierir. Kegen zerindeki box plotlar, ilgili deikenin zet istatistiklerini gstermektedir. Seilen Portland renk dzeni ise grafiin daha anlalr ve okunabilir ekilde olmasn salamaktadr.

```python
# 1. Plotly'nin figure_factory modln ie aktarr.
import plotly.figure_factory as ff

# 2. 2015 ylna ait verileri dataframe deikenine atar.
dataframe = timesData[timesData.year == 2015]

# 3. dataframe deikeninden belirli stunlar seer ve data2015 deikenine atar.
data2015 = dataframe.loc[:,["research","international", "total_score"]]

# 4. data2015 DataFrame'ine yeni bir "index" stunu ekler ve bu stuna ardk saylar atar.
data2015["index"] = np.arange(1,len(data2015)+1)

# 5. data2015 verilerini kullanarak bir scatter plot matrisi oluturur.
fig = ff.create_scatterplotmatrix(data2015, diag='box', index='index',colormap='Portland',
                                  colormap_type='cat', height=700, width=700)

# 6. Oluturulan scatter plot matrisini izer.
iplot(fig)
```

### - Animation with Plotly
Malum Plotlynin etkileimli grafikler oluturmak iin kullanlan popler bir ktphane olduunu belirtmitim. Dolaysyla Plotly, animasyonlar ve veri grselletirmelerine zaman boyunca dinamik bir boyut eklemek iinde kullanlabilmektedir.

zellikle Plotlynin animasyon kullanm alanlar:

* Zamanla Deien Veriyi Gstermek: Animasyonlar, zaman serisi verileri gibi zamanla deien veriyi grselletirmek iin idealdir. rnein, bir lkenin yllar boyunca GDPsinin nasl deitiini gsteren bir animasyon oluturabilirsiniz.

* Trendleri ve Desenleri Anlamak: Animasyon, verideki trendleri ve desenleri belirlemeyi ve anlamay kolaylatrabilir.

* Etkileyici Sunumlar: Animasyonlar, sunumlarnza veya web sitenize dinamik ve etkileyici grselletirmeler ekleyerek izleyicinizin dikkatini ekebilir.

* Senaryo Karlatrmalar: Farkl senaryolar veya durumlar karlatrmak iin animasyonlar kullanabilirsiniz. rnein, farkl senaryolar altnda bir simulasyonun sonularn gstermek.

Plotly ile animasyon kullanrken dikkate alnmas gereken hususlar ise;

* Bilgi Yklenmesi: ok fazla animasyon, izleyici zerinde bilgi yklenmesine neden olabilir. zellikle karmak veri setleriyle alrken animasyonlar basit ve anlalr tutmak nemlidir.

* Performans Sorunlar: Byk veri setleriyle yaplan animasyonlar, bazen yava yklenebilir veya taraycda taklabilir. Veri optimizasyonu yaparak bu tr sorunlarn nne gemek mmkndr.

* Uygun Aralk: Animasyonun ok hzl veya yava olmamasna dikkat edilmelidir. Animasyonun hz, izleyicinin veriyi rahata takip edebilmesine uygun olmaldr.

zetle, zamanla deien veriyi grselletirmek, trendleri ve desenleri anlamak, etkileyici sunumlar yapmak ve farkl senaryolar karlatrmak iin Plotly ile animasyonlar kullanlabilmektedir.

Plotly kullanarak izeceimiz bu son grafikle yeni bir data seti ile kii bana den gelir (GDP per Capita) ile yaam beklentisi (Life Expectancy) arasndaki ilikinin, yllar boyunca nasl deitiini gsteren bir scatter grafii oluturacaz.

* Grafikteki Her Nokta: Bir lkeyi temsil eder.

* Noktalarn Rengi: Farkl ktalar temsil eder. Renkler kta baznda ayrlmtr, bylece ayn ktadan gelen lkeler benzer renkte gsterilir.

* Noktalarn Boyutu: lkenin nfusunu temsil eder. Daha byk noktalar, daha yksek nfusa sahip lkeleri gsterirken, daha kk noktalar daha dk nfusa sahip lkeleri gsterir.

* Animasyon: Grafiin alt ksmnda bir zaman ubuu bulunmaktadr. Bu zaman ubuunu kullanarak yllar boyunca kii bana den gelirin ve yaam beklentisinin nasl deitiini gzlemleyebilirsiniz.

* Hover zellii: Herhangi bir noktann zerine geldiinizde, o noktann hangi lkeyi temsil ettiini grebilirsiniz.

zetle, bu grafik, yllar iindeki ekonomik bymenin (kii bana den gelir) yaam beklentisi zerindeki etkisini gsteren dinamik ve interaktif bir grselletirmedir. Ayrca ktalararas ve lkeleraras farkllklar da aka grebilirsiniz.

```python
import plotly.express as px  # Plotly Express ktphanesini ie aktarr.
df = px.data.gapminder()  # Gapminder veri setini df deikenine ykler.

# px.scatter fonksiyonunu kullanarak bir scatter grafii oluturur. 
# Bu grafikte x ekseninde kii bana den gelir (GDP per Capita), 
# y ekseninde ise yaam beklentisi (Life Expectancy) bulunmaktadr.
# animation_frame ile yllara gre animasyon eklenirken, 
# animation_group ile lkeler animasyon gruplar olarak belirlenmitir.
# size parametresi ile noktalarn boyutu lkenin nfusuna gre ayarlanmtr.
# color parametresi ile noktalar ktalarna gre renklendirilmitir.
# hover_name ile fareyle zerine gelindiinde lkenin ad gsterilmitir.
# log_x=True ile x ekseni iin logaritmik lek kullanlmtr.
# size_max=55 ile noktann maksimum boyutu belirlenmitir.
# range_x ve range_y ile x ve y eksenleri iin deer aralklar belirlenmitir.
px.scatter(df, 
           x="gdpPercap",
           y="lifeExp",
           animation_frame="year",
           animation_group="country",
           size="pop",
           color="continent",
           hover_name="country",
           log_x=True,
           size_max=55,
           range_x=[100,100000],
           range_y=[25,90])
```
